```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '../data/raw/fake_news/train.tsv'

columns = ['id', 'label', 'statement', 'subject', 'speaker', 'job', 'state', 'party', 
           'barely-true', 'false', 'half-true', 'mostly-true', 'pants-on-fire', 'context']

df = pd.read_csv(path, sep='\t', names=columns)

print("Перші 5 рядків даних:")
display(df.head())
```

```python
print("\nКількість прикладів по кожному класу:")
print(df['label'].value_counts())

sns.countplot(x='label', data=df)
plt.title('Розподіл класів (Fake News)')
plt.show()
```

```python
df['title_length'] = df['statement'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.histplot(df['title_length'], bins=30, kde=True)
plt.title('Гістограма довжини заголовків')
plt.xlabel('Кількість слів')
plt.show()
```
